```python
# Import necessary libraries

import pandas as pd
import json


# Load file

school_data = pd.read_csv("edubasealldata20251031-utf8.csv", low_memory=False, dtype=str)
pd.options.display.max_columns = None
school_data.head()
```

```python
# TypeOfEstablishment Mapping

typemap = school_data[["TypeOfEstablishment (code)", "TypeOfEstablishment (name)"]].drop_duplicates().sort_values("TypeOfEstablishment (code)")
typemap
```

```python
# EstablishmentStatus Mapping

statusmap = school_data[["EstablishmentStatus (code)", "EstablishmentStatus (name)"]].drop_duplicates().sort_values("EstablishmentStatus (code)")
statusmap
```

```python
# Convert TypeOfEstablishment & EstablishmentStatus to dictionaries (code: name)

type_dict = dict(zip(
    typemap["TypeOfEstablishment (code)"],
    typemap["TypeOfEstablishment (name)"]
))

status_dict = dict(zip(
    statusmap["EstablishmentStatus (code)"],
    statusmap["EstablishmentStatus (name)"]
))


# Combine into a single structure

combined = {
    "educational-establishment-type": type_dict,
    "educational-establishment-status": status_dict
}
combined
```

```python
# Write to JSON file

with open("../data/school-code-mappings.json", "w", encoding="utf-8") as f:
    json.dump(combined, f, indent=4, ensure_ascii=False)

print("JSON file saved as 'school-code-mappings.json'")
```
